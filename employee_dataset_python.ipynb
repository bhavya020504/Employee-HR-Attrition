```python
df["Gender"]=df["Gender"].astype("category")
```

```python
df["Overtime"]=df["Overtime"].astype("category")
```

```python
df["Attrition"]=df["Attrition"].astype("category")
df["Depatment"]=df["Department"].astype("category")
df.dtypes
```

```python
numeric_values=["Age","Salary","PerformanceScore","Experience"]
df[numeric_values].apply(pd.to_numeric)
df.dtypes
```

```python
df["EmployeeID"]=df["EmployeeID"].astype("string")
```

```python
df.dtypes
```

```python
salary_array = df["Salary"].to_numpy()
print("Average Salary (NumPy):", np.mean(salary_array))
print("Median Salary (NumPy):", np.median(salary_array))

```

```python
plt.hist(df["Salary"], bins=5, color="skyblue", edgecolor="black")
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Number of Employees")
plt.show()
```

```python
df_yes = df[df["Attrition"] == "Yes"]
df_no = df[df["Attrition"] == "No"]

plt.scatter(df_yes["WorkLifeBalance"], df_yes["Salary"], color="red", label="Attrition=Yes")
plt.scatter(df_no["WorkLifeBalance"], df_no["Salary"], color="green", label="Attrition=No")
plt.xlabel("WorkLifeBalance")
plt.ylabel("Salary")
plt.title("WorkLifeBalance vs Salary (Attrition)")
plt.legend()
plt.show()


```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("employee_dataset (2).csv")

```

```python
binary_cols=["Attrition","Overtime"]
for col in binary_cols:
  df[col]=df[col].map({"Yes":1,"No":0})
```

```python
df
```

```python
pd.get_dummies(df,columns=["Department"],drop_first=True)
```

```python

dept_cols = [col for col in df.columns if col.startswith('Department_')]

df[dept_cols] = df[dept_cols].astype(int)


print(df.head())
```

```python

df = pd.get_dummies(df, columns=['Department'], drop_first=True)


print(df.head())
```

```python

df.to_csv("employee_dataset_cleaned.csv", index=False)
```
